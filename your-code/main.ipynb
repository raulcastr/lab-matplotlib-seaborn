#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.

* Explore plt.xticks

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

fit_months = fitbit.groupby("Months_encoded")["Steps"].median()

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

fit_work = fitbit.groupby("Work_or_Weekend")["Steps"].median()
```

```python
fig3, axes3 = plt.subplots(1,2,figsize=(12,4))
plt.sca(axes3[0])
axes3[0].plot(fit_months.index, fit_months)
plt.xticks(range(13), ["","Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],rotation= "vertical")

plt.sca(axes3[1])
axes3[1].bar(fit_work.index, fit_work)
plt.xticks(range(2), ["Weekend","Workdays"])

#sns.barplot(data = fit_months.reset_index(), x = 'Months_encoded' , y = 'Steps' , ax = axes3[0] ) 
#sns.barplot(data = fit_work.reset_index(), x = 'Work_or_Weekend' , y = 'Steps' , ax = axes3[1] )
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
fitbit.head(3)
```

```python
# your code here

features=["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary"]
i = 0

fig4 , axes4 = plt.subplots(1, 3, figsize = (15,7))
fig4.suptitle('3 scatter plots', fontsize = 20)

for feature in features:
    
    axes4[i].scatter(fitbit[feature], fitbit["Steps"],  label = feature)
    axes4[i].set_title(feature)
    i+=1

```
